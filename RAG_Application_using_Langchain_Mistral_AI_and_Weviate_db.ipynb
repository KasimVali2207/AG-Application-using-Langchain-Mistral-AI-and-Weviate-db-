```python
from langchain.vectorstores import Weaviate
import weaviate

# Make sure these variables are set correctly
WEAVIATE_URL = "URL"  # Replace with your Weaviate URL
WEAVIATE_API_KEY = "Key"  # Replace with your actual API key

client = weaviate.Client(
    url=WEAVIATE_URL,
    auth_client_secret=weaviate.auth.AuthApiKey(WEAVIATE_API_KEY)  # Use weaviate.auth.AuthApiKey for authentication
)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
model_kwargs = {"device": "cpu"}
embedding_model_name = "sentence-transformers/all-mpnet-base-v2"
embeddings = HuggingFaceEmbeddings(model_name=embedding_model_name)
```

```python
!pip install pypdf
from langchain.document_loaders import PyPDFLoader
# PyPDFLoader is now in langchain_community
loader = PyPDFLoader("/content/RAG.pdf", extract_images=True)
pages = loader.load()
```

```python
pages
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
docs = text_splitter.split_documents(pages)
```

```python
docs
```

```python
vector_db = Weaviate.from_documents(
    docs, embeddings, client=client, by_text=False)
```

```python
vector_db.similarity_search("What is RAG?", k=3)
```

```python
vector_db.similarity_search("What is RAG?", k=3)[0].page_content
```

```python
print(
    vector_db.similarity_search(
        "What is attention?", k=3)
)
```

```python
from langchain.prompts import PromptTemplate
templates="""
Use the following pieces of context to answer the question at the end.
If you don't know the answer, just say that you don't know, don't try to make up an answer.
use ten sentences maximum and keep answer concise
context:{context}
Question: {question}
Helpful Answer:
"""
# Use PromptTemplate instead of ChatPromptTemplate
prompt = PromptTemplate.from_template(templates)
```

```python
from langchain import HuggingFaceHub
```

```python
from google.colab import userdata
HuggingfaceHub_api_token=userdata.get("HUGGINGFACE_TOKEN")
```

```python
model =HuggingFaceHub(
    huggingfacehub_api_token=HuggingfaceHub_api_token,
    repo_id="mistralai/Mistral-7B-Instruct-v0.1", # Corrected repo_id
    model_kwargs={"temperature":1, "max_length":180}
)
```

```python
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser
```

```python
output_parser=StrOutputParser()
```

```python
retriever=vector_db.as_retriever()
```

```python
rag_chain = (
    {"context": retriever,"question": RunnablePassthrough()}
     |prompt
     |model
     |output_parser
)
```

```python
rag_chain.invoke("What is RAG?")
```
